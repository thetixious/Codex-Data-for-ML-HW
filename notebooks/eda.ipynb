# EDA: Unified Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
df = pd.read_csv(r'data/raw/collected_unified.csv')
df.head()
```

## Class Distribution

```python
ax = df['label'].value_counts().plot(kind='bar', figsize=(8,4), title='Class Distribution')
ax.set_xlabel('label')
ax.set_ylabel('count')
plt.tight_layout()
```

## Text Length Distribution

```python
txt = df['text'].fillna('')
lengths = txt.str.len()
lengths.describe()
```

```python
lengths.hist(bins=40, figsize=(8,4))
plt.title('Text length distribution')
plt.xlabel('chars')
plt.ylabel('count')
plt.tight_layout()
```

## Top-20 Words

```python
words = ' '.join(df['text'].fillna('').astype(str)).lower().split()
top20 = Counter([w for w in words if len(w) > 2]).most_common(20)
pd.DataFrame(top20, columns=['word','count'])
```
